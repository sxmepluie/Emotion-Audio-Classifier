```python
import os
import yt_dlp
from google.colab import files

def video_indir_ve_sese_donustur(video_url, output_dir):



    # VİDEO İNDİR



    if not os.path.exists(output_dir):
        os.makedirs(output_dir, exist_ok=True)
        print(f"{output_dir} klasörü oluşturuldu.")

    ydl_opts = {
        'format': 'bestaudio/best',  # En iyi ses formatını seç
        'postprocessors': [{
            'key': 'FFmpegExtractAudio',  # Sadece ses çıkar
            'preferredcodec': 'mp3',  # MP3 formatında çıktı
            'preferredquality': '192',  # 192 kbps kalite
        }],
        'outtmpl': f'{output_dir}/%(title)s.%(ext)s',  # Dosya ismini video başlığına göre ayarla
    }

    try:
        with yt_dlp.YoutubeDL(ydl_opts) as ydl:
            ydl.download([video_url])  # Video URL'sinden ses indir
        print(f"Video başarıyla indirildi ve {output_dir} dizinine kaydedildi.")
    except Exception as e:
        print(f"İndirme sırasında bir hata oluştu: {e}")



def main():
    video_url = 'https://www.youtube.com/watch?v=jez-6Sw2lnw'  # Buraya video URL'sini girin
    output_dir = '/content/downloads'

    # Video indirme ve sese dönüştürme
    video_indir_ve_sese_donustur(video_url, output_dir)


if __name__ == "__main__":
    main()
```

```python
code = """
import os
import shutil
from pydub import AudioSegment
from pydub.silence import detect_nonsilent

def find_empty_audio_files(input_folder, output_folder, silence_threshold=-40, min_silence_len=1000):
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    remaining_files = []

    for filename in os.listdir(input_folder):
        if filename.endswith(".wav"):
            file_path = os.path.join(input_folder, filename)
            audio = AudioSegment.from_file(file_path)

            nonsilent_parts = detect_nonsilent(audio, min_silence_len=min_silence_len, silence_thresh=silence_threshold)

            if not nonsilent_parts:
                print(f"Moving silent file: {filename}")
                shutil.move(file_path, os.path.join(output_folder, filename))
            else:
                remaining_files.append(file_path)

    return remaining_files

def move_files_to_folder(files, destination_folder):
    if not os.path.exists(destination_folder):
        os.makedirs(destination_folder)
    for file_path in files:
        shutil.move(file_path, os.path.join(destination_folder, os.path.basename(file_path)))

folder_path = "neutral-datas"
output_folder = "bos_sesler"

remaining_files = find_empty_audio_files(folder_path, output_folder)

non_empty_folder = "konusma_sesleri"
move_files_to_folder(remaining_files, non_empty_folder)

print("İşlem tamamlandı.")
"""

# Dosyayı content/downloads klasörüne kaydetme
file_path = '/content/downloads/ses_isleme.py'
with open(file_path, 'w') as file:
    file.write(code)

print(f"Python kodu {file_path} dosyasına kaydedildi.")
```

```python
import os
import uuid
from pydub import AudioSegment



# SESLERİ PARÇALARA BÖL




def ses_dosyasini_bol(output_dir, parca_uzunlugu=30):
    mp3_dosyalar = [f for f in os.listdir(output_dir) if f.endswith('.mp3')]

    if not mp3_dosyalar:
        print("Hiçbir .mp3 dosyası bulunamadı.")
        return []

    input_file = os.path.join(output_dir, mp3_dosyalar[0])
    print(f"{input_file} dosyası işleniyor.")

    ses = AudioSegment.from_mp3(input_file)
    parca_uzunlugu_ms = parca_uzunlugu * 1000  # Saniyeden milisaniyeye dönüşüm
    parcalar = [ses[i:i + parca_uzunlugu_ms] for i in range(0, len(ses), parca_uzunlugu_ms)]

    parca_dosyalar = []
    for index, parca in enumerate(parcalar):
        benzersiz_ad = str(uuid.uuid4())
        chunk_file_path = os.path.join(output_dir, f"parca_{index + 1}_{benzersiz_ad}.mp3")
        parca.export(chunk_file_path, format="mp3")
        parca_dosyalar.append(chunk_file_path)
        print(f"{chunk_file_path} olarak kaydedildi.")

    return parca_dosyalar

output_dir = '/content/downloads'
os.makedirs(output_dir, exist_ok=True)
parca_dosyalar = ses_dosyasini_bol(output_dir)
```

```python
import os
import uuid
from pydub import AudioSegment
from spleeter.separator import Separator




# SESLERİ VOCALLERE AYIR




def ses_dosyasini_temizle(chunk_file, output_dir):
    separator = Separator('spleeter:2stems')
    try:
        separator.separate_to_file(chunk_file, output_dir)
    except Exception as e:
        print(f"Spleeter işleminde hata oluştu: {e}")
        return

    vocal_path = chunk_file.replace('.mp3', '/vocals.wav')

    if os.path.exists(vocal_path):
        ses = AudioSegment.from_wav(vocal_path)
        parca_suresi_ms = 5 * 1000
        parcalar = [ses[i:i + parca_suresi_ms] for i in range(0, len(ses), parca_suresi_ms)]

        for i, parca in enumerate(parcalar):
            benzersiz_ad = str(uuid.uuid4())
            parca_yolu = os.path.join(output_dir, f"vocal_parca_{i + 1}_{benzersiz_ad}.wav")
            parca.export(parca_yolu, format="wav")
            print(f"{parca_yolu} olarak kaydedildi.")

        os.remove(vocal_path)

downloads_path = '/content/downloads'

parca_dosyalar = [os.path.join(downloads_path, f) for f in os.listdir(downloads_path) if f.startswith('parca_') and f.endswith('.mp3')]

parca_dosyalar = sorted(parca_dosyalar, key=lambda f: int(f.split('_')[1].split('.')[0]))

output_dir = '/content/downloads/output'

for parca in parca_dosyalar:
    ses_dosyasini_temizle(parca, output_dir)
```

```python
import os
import shutil
from pydub import AudioSegment



# 5ER SANİYELERE AYIR VE ZİPLE




def vocal_dosyalarini_tasi_ve_ziple(source_dir, target_dir):
    # Hedef klasörü oluştur (varsa mevcut olanı silip yeniden oluştur)
    if os.path.exists(target_dir):
        shutil.rmtree(target_dir)
    os.makedirs(target_dir, exist_ok=True)

    # 'parca_' ile başlayan klasörleri alıyoruz
    for folder_name in sorted(os.listdir(source_dir)):
        if folder_name.startswith('parca_'):
            folder_path = os.path.join(source_dir, folder_name)

            # Her klasördeki vokal dosyasını buluyoruz
            for file_name in os.listdir(folder_path):
                if file_name.startswith('vocal'):
                    source_path = os.path.join(folder_path, file_name)
                    ses = AudioSegment.from_wav(source_path)

                    # 30 saniyelik bir ses dosyasını 5 saniyelik parçalara ayırma
                    parca_suresi_ms = 5 * 1000  # 5 saniye
                    parcalar = [ses[i:i + parca_suresi_ms] for i in range(0, len(ses), parca_suresi_ms)]

                    # Parçaları hedef dizine kaydediyoruz
                    for i, parca in enumerate(parcalar):
                        benzersiz_ad = f"{folder_name}_parca_{i + 1}.wav"
                        parca_yolu = os.path.join(target_dir, benzersiz_ad)
                        parca.export(parca_yolu, format="wav")
                        print(f"{parca_yolu} olarak kaydedildi.")

    # target_dir klasörünü zip dosyasına dönüştür
    zip_path = f"{target_dir}.zip"
    shutil.make_archive(target_dir, 'zip', target_dir)
    print(f"{target_dir} klasörü {zip_path} olarak ziplendi.")

    # output klasöründeki tüm dosyaları sil
    for folder_name in os.listdir(source_dir):
        folder_path = os.path.join(source_dir, folder_name)
        if os.path.isdir(folder_path):
            shutil.rmtree(folder_path)
            print(f"{folder_path} klasörü silindi.")
        else:
            os.remove(folder_path)
            print(f"{folder_path} dosyası silindi.")

# Kullanım için
source_dir = '/content/downloads/output'  # Ses dosyalarının bulunduğu klasör
target_dir = '/content/last_pieces'  # 5 saniyelik parçalara ayırıp taşıyacağımız hedef klasör
vocal_dosyalarini_tasi_ve_ziple(source_dir, target_dir)
```

```python
import os
import re
import shutil



# EN DÜŞÜK ... DOSYAYI SİL




downloads_path = '/content/downloads'

# Dosyaları al
files = os.listdir(downloads_path)

# Sayıları içeren dosyaları filtrele
def extract_number(file_name):
    match = re.search(r'parca_(\d+)', file_name)
    return int(match.group(1)) if match else float('inf')  # Sayı yoksa sonsuz kabul edilir

# Dosyaları sıralamak
files_with_numbers = [(file, extract_number(file)) for file in files if extract_number(file) != float('inf')]
files_with_numbers.sort(key=lambda x: x[1])

# En düşük 12 dosyayı veya dizini seçmek
files_to_delete = files_with_numbers[:12]

# Dosyaları veya dizinleri silme işlemi
for file, number in files_to_delete:
    file_path = os.path.join(downloads_path, file)
    if os.path.exists(file_path):
        if os.path.isdir(file_path):  # Eğer bu bir dizinse
            shutil.rmtree(file_path)  # Dizin ve içeriği silinir
            print(f"{file} dizini silindi.")
        else:  # Eğer dosya ise
            os.remove(file_path)
            print(f"{file} dosyası silindi.")
    else:
        print(f"{file} bulunamadı.")
```
